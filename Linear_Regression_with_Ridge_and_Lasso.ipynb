```python
#house Pricing Datasets

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
from sklearn.datasets import fetch_california_housing


housing = fetch_california_housing(as_frame=True)
data = housing.frame
target = housing.target

display(data.head())
display(target.head())
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
lin_reg=LinearRegression()
scores=cross_val_score(lin_reg,data,target,scoring="neg_mean_squared_error",cv=10)
print(np.mean(scores))
```

```python
#Ridge Regression
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
params={"alpha":[1e-15,1e-10,1e-29,1e-2,2,3,4,5,6]}
ridge=Ridge()
ridge_reg=GridSearchCV(ridge,params,scoring="neg_mean_squared_error",cv=10)
ridge_reg.fit(data,target)
print(ridge_reg.best_params_)
print(ridge_reg.best_score_)
```

```python
from sklearn.linear_model import Lasso
lasso=Lasso()
lasso_reg=GridSearchCV(lasso,params,scoring="neg_mean_squared_error",cv=10)
lasso_reg.fit(data,target)
print(lasso_reg.best_params_)
print(lasso_reg.best_score_)
```
